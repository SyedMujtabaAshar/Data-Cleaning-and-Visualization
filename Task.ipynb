## Steps to do: 3 - Visualizations

```python
#bar chart for one categorical variable

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt


# Bar chart for categorical data ('Sex = male or female ')
plt.figure(figsize=(8, 4))
data['Sex'].value_counts().plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Bar Chart of Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for numerical data ('Age')
plt.figure(figsize=(8, 4))
data['Age'].plot(kind='hist', bins=10, color='lightgreen', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate correlation matrix

# Select only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.1)
plt.title('Correlation Heatmap')
plt.show()
```

## Summary of Insights

1-Survival Rates by Class:
The bar chart shows that passengers in first class had a higher survival
rate compared to those in second and third class.

2-Age Distribution:
The histogram indicates that a majority of passengers were between 20 and
40 years old, with fewer children and older adults.

3-Correlation Insights:
The heatmap reveals a strong negative correlation between Pclass and Survived,
indicating that higher-class passengers were more likely to survive. There’s 
also a positive correlation between SibSp and Parch, suggesting that families
